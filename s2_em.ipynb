```python
class Point:
  def __init__(self, x, y):
    self.x = x
    self.y = y

class ColorPoint(Point):
  #color point class inheriting from point class
  def __init__(self,x,y,color):
    self.x= x
    self.y= y
    self.color= color

  def __str__(self):
    return f"{self.color}: P<{self.x},{self.y}>"

  def __lt__(self, other):
    # Compare ColorPoint objects based on their x-coordinate
    return self.x < other.x

p1= ColorPoint(1,2,"red")
p2= ColorPoint(3,4, "green")
p3= ColorPoint(5,6, "blue")
p4= ColorPoint(-2,-3,"maroon")
color_points= [p1,p2,p3,p4]
print(p1)
print(f"Unsorted ColorPoints: {[str(p) for p in color_points]}")
color_points.sort()
print(f"Sorted ColorPoints: {[str(p) for p in color_points]}")
```
